## Abordagens

Existem várias abordagens para se tratar a probabilidade: clássica,
frequentista, geométrica, subjetiva e axiomática. Nesta disciplina, estamos mais interessados na visão clássica e na visão frequentista:

1. **Visão Clássica**: a probabilidade é calculada com base em situações onde todos os resultados possíveis são conhecidos. Por exemplo, lançar um dado justo, onde cada face tem uma probabilidade de 1/6 de sair. Nesta visão, a probabilidade depende da quantidade de casos favoráveis e da quantidade total de casos.

2. **Visão Frequentista**: a probabilidade é baseada em observações de frequências reais de eventos em um grande número de repetições de um experimento. A probabilidade é a proporção do número de vezes que um evento ocorre em relação ao número total de tentativas.

Essas duas visões se completam em vários sentidos, sendo que o resultado de uma pode ser estimado em função da outra. Nesta disciplina, seguiremos um pouco mais adiante e definiremos variável aleatória.

## Variáveis Aleatórias

Uma variável aleatória em probabilidade é  uma maneira de atribuir números a eventos incertos. Em outras palavras, uma variável aleatória nada mais é que uma função que mapeia os resultados de um processo estocástico (aleatório) em valores numéricos.

Vamos usar como exemplo o lançamento de um dado. Suponha que você queira saber quantas vezes vai sair um número ímpar, como 1, 3 ou 5, no lançamento de dois dados. O processo estocástico é o lançamento desses dados e a variável aleatória $f$ é o modo como o mapeamento será feito em número:

**$f$** | **1** | **2** | **3** | **4** | **5** | **6**
------------|-------------|-------------|-------------|-------|----|-----
**1** | 2 | 1 | 2 | 1 | 2 | 1
**2** | 1 | 0 | 1 | 0 | 1 | 0
**3** | 2 | 1 | 2 | 1 | 2 | 1
**4** | 1 | 0 | 1 | 0 | 1 | 0
**5** | 2 | 1 | 2 | 1 | 2 | 1
**6** | 1 | 0 | 1 | 0 | 1 | 0

As variáveis aleatórias podem ser classificadas como discretas ou contínuas.

* **Discretas**: os resultados são frutos de contagem. Por exemplo, número de filhos, número de sucessos em $n$
 tentativas, e assim por diante.

* **Contínuas**: os resultados são frutos de mensuração. Por exemplo, massa (kg) do objeto , altura (m) do indivíduo,  temperatura, diâmetro de uma peça, e assim por diante.



## Valor Esperado

O valor esperado, também conhecido como esperança matemática, é um conceito importante na estatística. Em termos técnicos, ele mede o valor médio de todos os resultados possívels em uma variável aleatória. Por exemplo, para $f$, teríamos que o valor médio seria:

$$\frac{1}{36}(2 + 1 + 2 + 1 + 2 + 1 + 1 + 0 + 1 + 0 + 1 + 0  + 2 + 1 + 2 + 1 + 2 + 1 + 1 + 0 + 1 + 0 + 1 + 0 + 2 + 1 + 2 + 1 + 2 + 1 + 1 + 0 + 1 + 0 + 1 + 0) = \frac{1}{36}(2 \times 9+18 \times 1+0\times 9 = \frac{36}{36} = 1$$


### Abordagem Clássica

Na abordagem clássica, o valor esperado é o resultado de uma soma ponderada, que associa a probabilidade de ocorrência de um determinado evento com o valor atribuído pela variável aleatória para aquele evento. Em outras palavras, para se obter o valor esperado podemos obter todas os resultados e computar a probabilidade de cada ocorrência. Dessa forma, podemos usar a fórmula:

$$\mathbb{E}(X) = \sum_{x \in X}  x \mathbb{P}(x) $$



Por exemplo, para a variável aleatória $f$ teríamos:


$$\mathbb{E}(f) = 2 \times \mathbb{P}(f = 2) + 1 \times \mathbb{P}(f = 1) +  0 \times \mathbb{P}(f = 0) $$

Para ocorrer $f = 2$, os dois dados precisam ser ímpares, sendo que apenas 3 faces em cada dado é ímpar. Em outras palavras:

* $f = 2 \rightarrow P_{dado1}(1,3,5) \times P_{dado2}(1,3,5) = \frac{3}{6}\times\frac{3}{6} = \frac{1}{4}$.

Para ocorrer $f = 0$, os dois dados precisam ser pares, sendo que apenas 3 faces em cada dado é ímpar. Em outras palavras:

* $f = 0 \rightarrow P_{dado1}(2,4,6) \times P_{dado2}(2,4,6) = \frac{3}{6}\times\frac{3}{6} = \frac{1}{4}$.

Para ocorrer $f = 1$, um dado precisa ser par e o outro precisa ser ímpar. Em outras palavras:

* $f = 1 \rightarrow P_{dado1}(1,3,5) \times P_{dado2}(2,4,6) +  P_{dado1}(2,4,6) \times P_{dado2}(1,3,5) = \frac{3}{6}\times\frac{3}{6} + \frac{3}{6}\times\frac{3}{6} = \frac{1}{2}$.


Dessa forma:

$$\mathbb{E}(f) = 2 \times \mathbb{P}(f = 2) + 1 \times \mathbb{P}(f = 1) +  0 \times \mathbb{P}(f = 0) = 2 \times \frac{1}{4} + 1 \times \frac{1}{2} + 0 \times \frac{1}{4} = 1$$

### Abordagem Frequentista

Na abordagem frequentista, podemos repetir os experimentos um grande número de vezes e computar a média dos valores assumidos pela variável aleatória.

Nesse caso, se queremos saber qual o valor esperado do lançamento de um dado, podemos fazer vários experimentos e esperar uma convergência, conforme o código a seguir:

```
## Como criaremos ensaios aleatórios, precisamos da biblioteca random
import random
```

```
## Altere o número de experimentos
## para ver o quão rápido a convergência
## ocorre.
num_experimentos = 10000000

soma_f = 0.0
for i in range(num_experimentos) :
    dado1 = random.randint(1,6)
    dado2 = random.randint(1,6)

    # A seguir somamos a quantidade de ímpares
    soma_f += (dado1 % 2 == 1) + (dado2 % 2 == 1)
soma_f/num_experimentos
```

Ao aumentar o número de experimentos, notamos que a média converge para o número 1, sendo então que este é o valor esperado pela nossa definição.

### Propriedades do Valor Esperado

1. Seja $X$ uma variável aleatória, \mathbb{E}[cX]=c\mathbb{E}[X]

    * Significa dizer que o valor esperado de uma constante vezes uma variável aleatória  é igual à constante vezes o valor esperado da variável aleatória.

2. Sejam $X$ e $Y$ variáveis aleatórias definidas sobre o mesmo grupo de eventos, $\mathbb{E}[X±Y]= \mathbb{E}[X] ± \mathbb{E}[Y]$

Esta segunda propriedade será melhor entendida como um exemplo. Acima usamos a função $f$ que contabilizava o número de elementos ímpares no lançamento de dois dados. Podemos dizer que o valor de $f$ pode ser definido como a soma do valor de duas outras variáveis aleatórias $g$ e $h$, tais que:

* $g$ é o número de ímpares no lançamento do primeiro dado.

* $h$ é o número de ímpares no lançamento do segundo dado.

Assim, $f = g + h$ e pela propriedade temos que:
$$\mathbb{E}(f) = \mathbb{E}(g+h) = \mathbb{E}(g) + \mathbb{E}(h)$$

É muito fácil computar o valor esperado de $g$ e $h$, pois só são possíveis os valores $1$ e $0$, sendo que em metade dos casos teremos um valor par e em metado dos casos teremos um valor ímpar. Assim.

$$\mathbb{E}(g) = \mathbb{E}(h) = 1 \times \frac{1}{2} + 0 \times \frac{1}{2} = \frac{1}{2}$$

Concluímos facilmente então que:

$$\mathbb{E}(f) = \mathbb{E}(g) + \mathbb{E}(h) = \frac{1}{2} + \frac{1}{2} = 1$$

Note que foi muito mais fácil computar o valor esperado de $f$ usando a propriedade da soma do valor esperado.

O valor final seria chamado de recompensa futura. Essa recompensa seria o quanto o dono do robô espera ganhar  por simplesmente deixar o robô parado em uma parte específica do jardim.

Um valor previsto para ganhar em algum momento no futuro é incerto, então um fator de desconto estima o ganho a longo prazo. Em outras palavras, um retorno que ganharemos daqui a quatro passos tem menos valor do que um retorno que ganharemos no próximo passo. Vamos considerar que o dono do robô estipulou um fator de desconto $\gamma = 0.7$.

## Abordagem Frequentista

Notemos que cada episódio fornece um valor final muito diferente. Vamos assumir que o robô trabalha apenas sete horas por dia. Na abordagem frequentista, precisamos repetir o experimento várias e várias vezes para depois computar a média dos valores retornados.

### Exercício:

Seu trabalho é gerar os episódios e ver o ganho obtido. O código a seguir já foi parcialmente iniciado e você deverá seguir programando o que precisa ser feito para depois computar a média.


```
#####
# Configurações dos experimento
#####

## Esse dicionário funciona da seguinte maneira.
## A chave é o ponto de partida do nosso agente
## em uma ação. O valor é um vetor que corresponde
## a todos os pontos para onde o robô pode se
## movimentar seguindo a critério: [left, up, right, down]
grid = {
    "A" : ["A", "A", "B", "C"],
    "B" : ["A", "B", "B", "D"],
    "C" : ["A", "C", "C", "D"],
    "D" : ["B", "C", "D", "D"]
}

## Esse dicionário mostra a recompensa por chegar a um estado
reward = {
    "A" : 0,
    "B" : 5,
    "C" : 0,
    "D" : 0
}

## Essa função move ao acaso no jardim, a partir
## de uma entrada.
def move(entrada) :
    return random.choice(grid[entrada])

## Essa função gera um episódio, que no caso seria
## o caminho do robô no jardim por um dia inteiro
## de trabalho.
def episodio(inicio = "A", num_passos = 24, gamma = 0.7) :
    posicao = inicio
    ganho_acumulado = 0
    for i in range(num_passos) :
        ## Todo: Complete o código fazendo
        ## um movimento para uma dada posição,
        ## coletando a recompensa e adicionando
        ## no ganho acumulado descontado por
        ## gamma
        posicao = move(posicao)
        ganho_acumulado += reward[posicao] * pow(gamma, i)

    return ganho_acumulado
```

```
## Altere o número de experimentos
num_experimentos = 100000
## Agora execute vários experimentos e veja o
## valor esperado segundo a abordagem frequentista.
for inicio in ["A", "B", "C", "D"] :
    soma = 0.0
    for i in range(num_experimentos) :
        soma += episodio(inicio)
    print("%s --> %.2f" % (inicio, soma/num_experimentos))
```

Ao executar os experimentos acima, notamos que em média estar em **A** nos fornece um ganho acumulado de aproximadamente **4.16**, sendo que o mesmo acontece para **D**. Também notamos que estar em **C** nos fornece um ganho acumulado de apenas **2.23**. Por fim, estar em **B** nos fornece um ganho acumulado de aproximadamente **6.0**.

## Abordagem Clássica

Resolver o cenário do robô jardineiro com a abordagem clássica será um desafio um pouco maior, então precisamos avançar com muita paciência. O que queremos obter é o valor esperado do ganho acumulado, sendo que esse ganho acumulado pode ser dado pela fórmula a seguir, onde $R_t$ é a recompensa obtida pelo robô na hora $t$:

$$G =  R_{1} + \gamma R_{2} + \gamma^2 R_{3} +  \ldots + \gamma^{23} R_{24} $$

Como $\gamma^{23}$ é um valor desprezível, por $\gamma$ ser menor do que 1, vamos simplificar os cálculos assumindo que G é uma soma infinita.

$$G =  R_{1} + \gamma R_{2} + \gamma^2 R_{3} +  \ldots  $$

Vamos, por fim, generalizar a fórmula acima para permitir computar o ganho acumulado inicianco em qualquer hora do dia.

$$G_t =  R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} +  \ldots $$

Em outras palavras, no tempo $t$, o robô está trabalhando em uma parte do jardim e fará um movimento para um estado adjacente que fornecerá um donativo $R_{t+1}$. A partir desse estado, os novos movimentos gerarão recompensas descontadas $R_{t+2}$, $R_{t+3}$, ...


Note que $G_t$ pode ser escrito como uma recorrência:

$$G_t =  R_{t+1} + \gamma \left[ R_{t+2} + \gamma R_{t+3} + \gamma^2 R_{t+4} +  \ldots \right]$$

$$ G_t = R_{t+1} + \gamma G_{t+1} $$


Esta recorrência é uma equação que você usará com muita frequência, garanta que ela foi compreendida com perfeição:


Vamos tentar computar o valor esperado quando o robô começa em uma parte específica do jardim que chamaremos de $s$, ou seja, queremos $\mathbb{E}(G_t | S_t = s)$, sendo que $s$ pode significar qualquer uma das partes $A$, $B$, $C$ e $D$.

$$ \mathbb{E}(G_t | S_t = s) = \mathbb{E}[ R_{t+1} + \gamma G_{t+1} | S_t = s]$$

Note que o valor que receberemos em $t+1$, ou seja, $R_{t+1}$ depende da próxima ação do robô. Como o robô só tem quatro ações a tomar, sendo todas elas equiprováveis, saberemos o valor de $R_{t+1}$ quando considerarmos todos os eventos a partir de $t$. Em outras palavras, usaremos a fórmula do valor esperado:

$$\mathbb{E}(X) = \sum_{x \in X}  x \mathbb{P}(x) $$

Onde $\mathbb{P}(X) = \frac{1}{4}$ para qualquer uma das direções tomadas pelo robô e chamaremos de $R_{t+1} = r$ para a recompensa ao se tomar aquela direção. Assim:


$$ \mathbb{E}(G_t | S_t = s) = \mathbb{E}[ R_{t+1} + \gamma G_{t+1} | S_t = s] = \sum_{a}\mathbb{P}(a|s) (r + \gamma \mathbb{E}(G_{t+1} | S_{t+1} = s'))$$

Nessa fórmula, o  estado $s'$ é aquele atingido após a ação $a$. O valor esperado do retorno acumulado nesse estado vai nos ajudar a saber o valor esperado em $s$.

Para a notação ficar mais fácil de acompanhar, vamos chamar de $V(s)$ o valor esperado em questão. Em outras palavras, $V(s) = \mathbb{E}[G_t | S_t = s]$.

A fórmula parece muito complicada, então vamos ganhar um mais pouco de intuição sobre o seu significado com  o exemplo que estamos trabalhando.

Vamos começar com o estado $A$, vamos supor que o dono do robô colocou ele nessa parte do jardim:

\begin{align}
V(s) = \mathbb{E}(G_t | S_t = A) & = \sum_{a}\mathbb{P}(a|A) (r + \gamma \mathbb{E}(G_{t+1} | S_{t+1} = s'))\\
           & = \sum_{a}\frac{1}{4}(r + 0.7 V(s'))\\
           & = \frac{1}{4}(5+0.7V(B))+\frac{1}{4}(0+0.7V(C)) + \frac{1}{4}(0+0.7V(A))+ \frac{1}{4}(0+0.7V(A))\\
           & = \frac{1}{4}(5+0.7V(B))+\frac{1}{4}(0+0.7V(C)) + \frac{1}{2}(0+0.7V(A))\\
\end{align}

Equações similares podem ser escritas para os outros estados:

\begin{align}
V(B) & = \frac{1}{2}(5+0.7V(B))+\frac{1}{4}(0+0.7V(A) + \frac{1}{4}(0+0.7V(D)) \\
V(C) & = \frac{1}{4}(0+0.7V(A))+\frac{1}{4}(0+0.7V(D) + \frac{1}{2}(0+0.7V(C)) \\
V(D) & = \frac{1}{4}(5+0.7V(B))+\frac{1}{4}(0+0.7V(C) + \frac{1}{2}(0+0.7V(D)) \\
\end{align}

Isso resulta em um sistema com quatro equações e quatro variáveis que pode ser resolvido manualmente ou com algum programa apropriado. As soluções para esse problema são:

\begin{align}
V(A) & = 4.2\\
V(B) & = 6.1\\
V(C) & = 2.2\\
V(D) & = 4.2
\end{align}

```
## Configurações do problema.
rewards = [0, 5, 0, 0] # A, B, C, D
policy  = [0.25, 0.25, 0.25, 0.25] # up, right down, left (moves probabilities)
gamma   = 0.7

current_values = [0,0,0,0] ## Variáveis para guardar os valores esperados a cada iteração
all_values     = [] ## Vetor para armazenar os state-values a cada iteração

for i in range(10):

  next_values    = [None, None, None, None]
  # State A
  next_values[0] = (policy[0]*(rewards[0]+gamma*current_values[0]) +
                    policy[1]*(rewards[1]+gamma*current_values[1]) +
                    policy[2]*(rewards[2]+gamma*current_values[2]) +
                    policy[3]*(rewards[0]+gamma*current_values[0]))

  # State B
  next_values[1] = (policy[0]*(rewards[1]+gamma*current_values[1]) +
                    policy[1]*(rewards[1]+gamma*current_values[1]) +
                    policy[2]*(rewards[3]+gamma*current_values[3]) +
                    policy[3]*(rewards[0]+gamma*current_values[0]))

  # State C
  next_values[2] = (policy[0]*(rewards[0]+gamma*current_values[0]) +
                    policy[1]*(rewards[3]+gamma*current_values[3]) +
                    policy[2]*(rewards[2]+gamma*current_values[2]) +
                    policy[3]*(rewards[2]+gamma*current_values[2]))

  # State D
  next_values[3] = (policy[0]*(rewards[1]+gamma*current_values[1]) +
                    policy[1]*(rewards[3]+gamma*current_values[3]) +
                    policy[2]*(rewards[3]+gamma*current_values[3]) +
                    policy[3]*(rewards[2]+gamma*current_values[2]))

  all_values.append([ [current_values[0],
                       current_values[1]],
                     [current_values[2], current_values[3]]   ])


  if (current_values == next_values) :
    break
  current_values = next_values
print(current_values)
```
